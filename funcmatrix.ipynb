```python
def matrix(Row,Column):
    Matrix=[]
    for i in range(Row):
        a=[]
        for j in range(Column):
            a.append(int(input("enter the number of elemets")))
        Matrix.append(a)
    return Matrix
```

```python
def add_matrix(matrix1,matrix2):
    row=len(matrix1)
    column=len(matrix1[0])
    result=[[0 for _ in range(column)]for _ in range(row)]
    for i in range(row):
        for j in range(column):
            result[i][j]=matrix1[i][j]+matrix2[i][j]
    return result
```

```python
matrix1=matrix(2,3)
matrix2=matrix(2,3)
print("matrix1:",matrix1)
print("matrix2:",matrix2)
```

```python
add=add_matrix(matrix1,matrix2)
print(add)
```
